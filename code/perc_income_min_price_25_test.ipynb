```python
MEASURE_NAME = 'perc_income_min_price_25'
```

```python
bbn_files = sorted(pathlib.Path(settings.BBN_DATA_DIR).glob("*.csv.xz"))
```

```python
acs_files = sorted(pathlib.Path(settings.ACS_DATA_DIR).glob("**/*.csv.xz"))
```

```python
overlapped_filenames = sorted(set([f.name for f in bbn_files]).intersection(set([f.name for f in acs_files])))
overlapped_filenames
```

```python
bbn_files = [f for f in bbn_files if f.name in overlapped_filenames]
acs_files = [f for f in acs_files if f.name in overlapped_filenames]
```

```python
bbn_df = pd.read_csv(bbn_files[0], dtype={'GEOID20':object})
bbn_df
```

```python
# Remove all not nulls but also the specific none-standard values
bbn_df = bbn_df[(bbn_df["price"] >= 0) & (bbn_df["price"].notnull())].copy()
bbn_df = bbn_df.rename(columns={"GEOID20": "geoid"})
bbn_df['speed_value'] = bbn_df['speed'].apply(lambda x: int(x.split(' Mbps')[0]))
bbn_df
```

```python
acs_df = pd.read_csv(acs_files[0], dtype={'GEOID21':object})
acs_df
```

```python
bbn_df_u_25 = bbn_df[(bbn_df['down_up']=='Upload') & (bbn_df['speed_value'] >= 25)]
bbn_df_u_25
```

```python
bdf = bbn_df_u_25.groupby(['geoid','year_parsed'])['price'].agg('min').to_frame().reset_index()
bdf = bdf.rename(columns={'year_parsed': 'year'})
bdf
```

```python
acs_df = acs_df.rename(columns={'GEOID21':'geoid'})
acs_df['year'] = 2021
adf = acs_df.groupby(['geoid','year'])['B19013_001E'].agg('min').to_frame().reset_index()
adf = adf[adf['B19013_001E'] > 0]
adf
```

```python
adf['year'] = 2023 # Future should be adjusted for inflation?
```

```python
fdf = pd.merge(adf, bdf, on=['geoid','year'])
fdf['measure'] = 'perc_income_min_price_25'
fdf['value'] = fdf['price']/fdf['B19013_001E'] * 100
fdf = fdf.reindex(settings.STANDARD_COLS, axis="columns")
fdf['region_type'] = 'block_group'
fdf
```

```python
def generate_higher_geo(df, geoid_length=11):
    region_type = settings.FIPS_LENGTH_DESIGNATION[geoid_length]
    df[region_type] = df["geoid"].str[:geoid_length]
    cdf = (
        df.groupby([region_type, "year"])["value"].agg("mean").to_frame().reset_index()
    )
    cdf["measure"] = MEASURE_NAME
    # rename the census tract back into geoid to concat
    cdf = cdf.rename(columns={region_type: "geoid"})
    cdf = cdf.reindex(settings.STANDARD_COLS, axis="columns")
    assert region_type not in cdf.columns
    cdf["region_type"] = region_type
    return cdf
```

```python
c_df = generate_higher_geo(fdf, 5)
c_df
```

```python
ct_df = generate_higher_geo(fdf, 11)
ct_df
```
